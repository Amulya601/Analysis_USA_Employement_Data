# Dataset1:

EMPLOYEMENT DATA WITH WAGES in USA:

    The dataset contains the wage mean and medians in various employement sectors as per survey from 2021.

DATASET LINK:

    https://www.bls.gov/oes/current/oes_nat.htm
    
COLUMNS IN DATASET:

Occupation_code  : A unique code associated with the each different occupation type.

Occupation_title : Lists out of the different types of occupations in USA by 2019

Group            : Level of their Occupation

Total_Employees  : Total Number of Employees

Hourly Mean      : Mean value of wages hourly

Annual_Mean      : Annual Mean wage

Hourly Median    : Hourly Median wage

ANNUAL           : Contains "TRUE" if only annual wages are released. The OEWS program releases only annual wages for some occupations that typically work fewer than 2,080 hours per year, but are paid on an annual basis, such as teachers, pilots, and athletes.

HOURLY           : Contains "TRUE" if only hourly wages are released. The OEWS program releases only hourly wages for some occupations that typically work fewer than 2,080 hours per year and are paid on an hourly basis, such as actors, dancers, and musicians and singers.

### Merging two datasets

Merging the two datasets with common column occupation title and using outer join to involve all the data.

```python
df=pd.merge(state_data,total_employement, on='OCC_TITLE', how='outer')
#displaying dataset after merge
df
```

Dropping the columns:
    
Reasons for dropping
AREA : Area code is not required as it doesn't hold any zip code.
HOURLY_x, ANNUAL_x, ANNUAL_y,HOURLY_y : Most of the data is NULL


```python
df=df.drop(['AREA','O_GROUP_y', 'HOURLY_x', 'ANNUAL_x', 'ANNUAL_y','HOURLY_y', 'OCC_CODE_y'], axis=1)
```

Renaming the columns for better readability and understanding.

```python
df=df.rename(columns={"OCC_CODE_x": "OCC_CODE", "AREA_TITLE": "STATE","O_GROUP_x":"GROUP","TOT_EMP_x":"STATE_EMP","H_MEAN_x":"STATE_H_MEAN","A_MEAN_x":"STATE_A_MEAN", "H_MEAN_y":"NATION_H_MEAN","TOT_EMP_y":"NATION_EMP","H_MEDIAN_x":"STATE_H_MEDIAN","H_MEDIAN_y":"NATION_H_MEDIAN","A_MEDIAN_x":"STATE_A_MEDIAN", "H_MEDIAN_Y":"NATION_MEDIAN_x", "A_MEAN_y":"NATION_A_MEAN", "A_MEDIAN_y":"NATION_A_MEDIAN"})
```

```python
df
```

```python
#replacing the symbols with null values
FinalData=df.replace('#', np.nan)
FinalData=FinalData.replace('*', np.nan)
FinalData=FinalData.replace('**', np.nan)
FinalData
```

```python
#Dropping the rows with null values
FinalData=FinalData.dropna()
FinalData
```
